```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
import os
from urllib.request import urlretrieve
import pandas as pd

URL = "https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD"

def get_fremont_data(filename='Fremont.csv', url=URL, force_download=False):
    if force_download or not os.path.exists(filename):
        urlretrieve(url, filename)
    data = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
    data.columns = ['West', 'East']
    data['Total'] = data['West'] + data['East']
    return data
```

```python
data = get_fremont_data()
data.head()
```

```python
data.resample('W').sum().plot();    # data.plot() had too many data points for the dataset to plot so resampling it into Weeks
```

```python
data.groupby(data.index.time).mean().plot();
```

```python
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.iloc[:5, :5]
```

```python
pivoted.plot(legend=False, alpha=0.01);
```

The two peaks that go up during the morning and the evening may show the weekdays and the faint broader lines may show the weekend communtes
